```python
def get_url(city):
    base = "https://www.comingsoon.it/cinema/"+ city+"/"
    return base 

def get_movie(div):
    nome_film = div.find('a','h5 titolo cat-hover-color anim25').text
    for ul in div('ul','col-xs-12 col-sm-9 box-descrizione'):
        descrizione = ul.find('li').text
        
        lis = []
        for li in ul.findAll('li'):
            lis.append(li)
       
    genere = lis[1].text
        
    return{
        "nome_film" : nome_film,
        "descrizione": descrizione,
        "genere": genere
    }


from bs4 import BeautifulSoup
import requests
from time import sleep


print("In che città vivi?")
city=input()
url = get_url(city)
films = []
soup = BeautifulSoup(requests.get(url).text, 'html5lib')

for div in soup('div', 'col-xs-12 cinema'):
    films.append(get_movie(div))

print ("Film trovati: ", len(films))
print("\n")

for film in films:
    print(film["nome_film"])
    print(film["descrizione"])
    print(film["genere"])
    print("\n")
```
